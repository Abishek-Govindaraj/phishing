```python
plt.figure(figsize=(14,10))
plt.subplot(2,2,1)
plt.plot(df['type'],df['Organization'],color="green")
plt.title("type vs Organization")
plt.xlabel("type")
plt.ylabel("Organization")
plt.tight_layout()
plt.subplot(2,2,2)
plt.hist(df['Organization'],color="orange",bins=4)
plt.title("type vs Organization")
plt.xticks(rotation=45,ha="right")
plt.tight_layout()
plt.subplot(2,2,3)
plt.bar(df['type'],df['Organization'],color="blue")
plt.title("type vs Organization")
plt.xlabel("type")
plt.ylabel("Organization")
plt.tight_layout()
plt.subplot(2,2,4)
plt.scatter(df['type'],df['Organization'],color="red")
plt.title("type vs Organization" )
plt.xlabel("type")
plt.ylabel("Oragination")
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.tight_layout()
plt.savefig("report.png")
plt.show()

```
